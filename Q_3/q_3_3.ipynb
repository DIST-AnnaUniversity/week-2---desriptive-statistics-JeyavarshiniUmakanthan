```python
import pandas as pd
import numpy as np
from google.colab import files
import io
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['Vehicles Registered.csv']))
```

```python
import numpy as np

cols = ['Newly_Registered_2016_17']
for col in cols:
  print("Column ", col)
  data = df[col]
  print(data)
  df.boxplot(cols,figsize=(10,15),fontsize='8',grid=False)
  print('\n')
  Q1 = np.percentile(data, 25, interpolation = 'midpoint')
  Q3 = np.percentile(data, 75, interpolation = 'midpoint')
  IQR = Q3 - Q1
  print('Q1 \tQ3 \tIQR')
  print(Q1,Q3,IQR)
  LQR = Q1-(1.5*IQR)
  UQR = Q3+(1.5*IQR)
  
  print('\n')
  print("Outliers are:")
  for i in data:
    if i < LQR or i > UQR:
      print(i)
  
```
